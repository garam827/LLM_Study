## LangChain Expression Language (LCEL)

LCEL은 LangChain에서 체인 구성 및 조작을 위한 선언적이고 유연한 방법을 제공하는 모듈입니다. 이는 LLM(대규모 언어 모델) 애플리케이션 개발을 간소화하고, 모듈화된 구성 요소들을 쉽게 조합하여 복잡한 워크플로우를 만들 수 있도록 돕습니다.

**주요 특징:**

*   **직관적인 구성:** 파이프(`|`) 연산자를 사용하여 구성 요소들을 연결하여 데이터 흐름을 시각적으로 쉽게 이해할 수 있습니다.
*   **비동기 지원:** 비동기 I/O를 기본적으로 지원하여 효율적인 병렬 처리가 가능합니다.
*   **스트리밍 지원:** 실시간 데이터 스트리밍을 지원하여 응답 시간을 단축하고 사용자 경험을 향상시킵니다.
*   **병렬 처리:** 여러 구성 요소를 동시에 실행할 수 있어 성능을 최적화합니다.
*   **강력한 타입 시스템:** 입력 및 출력에 대한 타입 힌트를 통해 코드의 안정성을 높이고 디버깅을 용이하게 합니다.
*   **쉬운 통합:** 기존 LangChain 구성 요소(프롬프트 템플릿, LLM, 출력 파서 등)와 원활하게 통합됩니다.

**LCEL을 사용하는 이유:**

LCEL은 복잡한 LLM 애플리케이션을 구축할 때 코드의 가독성을 높이고, 재사용성을 향상시키며, 유지보수를 용이하게 합니다. 또한, 비동기 및 스트리밍 기능을 통해 성능과 사용자 경험을 크게 개선할 수 있습니다.

```python
import os
from google.colab import userdata

API_KEY = userdata.get('OPEN_ROUTER')
os.environ['OPENAI_API_KEY'] = API_KEY
```

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model_name = 'nvidia/nemotron-3-super-120b-a12b:free', openai_api_base = 'https://openrouter.ai/api/v1')

from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content = "너는 미녀와 야수에 나오는 미녀야. 그 캐릭터에 맞게 사용자와 대화하라."),
    HumanMessage(content = "안녕? 저는 개스톤입니다. 오늘 시간 괜찮으시면 저녁 같이 먹을까요?"),
]

model.invoke(messages)
```

```python
# 텍스트 결과만 추출

from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

result = model.invoke(messages)
parser.invoke(result)
```

```python
# Langchain | 연산자

chain = model | parser
chain.invoke(messages)
```

## 프롬프트 템플릿 이용

```python
from langchain_core.prompts import ChatPromptTemplate

system_template = "너는 {story}에 나오는 {character_a} 역할이다. 그 캐릭터에 맞게 사용자와 대화하라."
human_template = "안녕? 저는 {character_b}입니다. 오늘 시간 괜찮으시면 {activity} 같이 할까요?"

prompt_template = ChatPromptTemplate([
    ("system", system_template),
    ("user", human_template),
])

result = prompt_template.invoke({
    "story": "미녀와 야수",
    "character_a": "미녀",
    "character_b": "야수",
    "activity": "저녁"
})

print(result)
```

```python
# 체인 구성
chain = prompt_template | model | parser

chain.invoke({
    "story": "미녀와 야수",
    "character_a": "미녀",
    "character_b": "야수",
    "activity": "저녁"
})
```

```python
chain = prompt_template | model | parser

chain.invoke({
    "story": "미녀와 야수",
    "character_a": "미녀",
    "character_b": "개스톤",
    "activity": "저녁"
})
```

```python
from typing import Literal
from pydantic import BaseModel, Field

class Adlib(BaseModel):
    """스토리 설정과 사용자 입력에 반응하는 대사를 만드는 클래스"""
    answer: str = Field(description = "스토리 설정과 사용자와의 대화 기록에 따라 생성된 대사")
    main_emotion: Literal["기쁨", "분노", "슬픔", "공포", "냉소", "불쾌", "중립"] = Field(description = "대사의 주요 감정")
    main_emotion_intensity: float = Field(description = "대사의 주요 감정의 강도 (0.0 ~ 1.0)")

structured_llm = model.with_structured_output(Adlib)
adlib_chain = prompt_template | structured_llm

adlib_chain.invoke({
    "story": "미녀와 야수",
    "character_a": "벨",
    "character_b": "개스톤",
    "activity": "저녁"
})
```
