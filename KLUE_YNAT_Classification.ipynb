## 3.3.2 토크나이저 활용하기
- 토크나이저는 텍스트를 토큰 단위로 나누고 각 토큰을 대응하는 토큰 아이디로 변환하다.
- 토크나이저도 학습 데이터를 통해 어휘사전을 구축 --> 일반적으로 모델과 함께 저장
- 허깅페이스 허브에서 모델과 토크나이저를 불러오는 경우 동일한 모델 아이디로 맞춰야 한다.

````python
# 토크나이저 불러오기 : klue/roberta-base 저장소의 토크나이저를 불러온다.
from transformers import AutoTokenizer

model_id    = 'klue/roberta-base'
tokenizer   = AutoTokenizer.from_pretrained(model_id)
````

1) input_ids
    - 토큰 아이디 리스트
    - 토큰화 했을 때 각 토큰의 토크나이저 사전의 몇 번째 항목인지를 나타냄.
2) attention_mask
    - 실제 텍스트인지 아니면 길이를 맞추기 위한 패딩인지 알려줌
    - 1이면 실제 토큰

3) token_type_ids
    - 토큰이 속한 문장의 아이디
    - 0이면 일반적으로 첫 번째 문장

- 토큰 아이디를 다시 텍스트로 돌리려면 토크나이저의 decode 메서드를 사용
- 특수 토큰 제외하려면 skip_special_tokens 인자를 True로 설정

````python
tokenized = tokenizer('토크나이저는 텍스트를 토큰 단위로 나눈다.')

print(tokenized)
print(tokenizer.convert_ids_to_tokens(tokenized['input_ids']))
print(tokenizer.decode(tokenized.input_ids))
print(tokenizer.decode(tokenized.input_ids, skip_special_tokens = True))
````

- 토크나이저는 한 번에 여러 문장을 처리할 수 있다.

````python
# 토크나이저에 여러 문장 넣기
tokenizer(['첫 번째 문장', '두 번째 문장'])
````

- 2개의 문장을 한번에 넣는 경우
    - 예) 2개 문장이 서로 원인/결과인지 학습
    - 이 때, 2개 문장의 하나의 데이터라는 것을 표시하기 위해 한번 더 리스트로 감싸준다.

````python
# 하나의 데이터에 여러 문장이 들어가는 경우
tokenizer([['첫 번째 문장', '두 번째 문장']])
````

## 3.4.1 데이터 준비
- KLUE 데이터셋 YNAT 서브셋 : 연합 뉴스 기사 제목 및 카테고리 정보

````python
from datasets import load_dataset

klue_tc_train   = load_dataset('klue', 'ynat', split = 'train')
klue_tc_eval    = load_dataset('klue', 'ynat', split = 'validation')

klue_tc_train
````

````python
# 실습에 사용하지 않는 불필요한 컬럼 제거
klue_tc_train   = klue_tc_train .remove_columns(['guid', 'url', 'date'])
klue_tc_eval    = klue_tc_eval  .remove_columns(['guid', 'url', 'date'])
````

````python
# 카테고리를 문자로 표기한 label_str 컬럼 추가
klue_tc_label = klue_tc_train.features['label']

# 데이터셋의 features에서 label 필드(보통 ClassLabel 타입)의 메타정보 객체를 꺼냅니다.
# 이 객체는 정수↔문자열 레이블 매핑(예: 0 → "sports", 1 → "politics" 등)을 가지고 있습니다.

def make_str_label(batch):
    batch['label_str'] = klue_tc_label.int2str(batch['label'])
    return batch

# 배치(예: 여러 샘플을 묶은 dict)를 입력받아 label_str이라는 새 키를 추가하는 함수입니다.
# klue_tc_label.int2str(...)는 정수 레이블(또는 정수 리스트)을 해당하는 문자열 레이블로 변환합니다.
# 변환된 문자열(혹은 문자열 리스트)을 batch['label_str']에 할당하고, 수정된 배치를 반환합니다.

klue_tc_train = klue_tc_train.map(make_str_label, batched = True, batch_size = 1000)
# map을 사용해 데이터셋의 모든 샘플에 대해 make_str_label을 적용합니다.
# batched=True라서 make_str_label에 들어오는 batch는 단일 샘플이 아니라 여러 샘플을 묶은 딕셔너리입니다.
# batch_size=1000은 한 번에 처리할 샘플 수(배치 크기)를 지정합니다.
# 결과로 label_str 컬럼이 추가된 새로운(또는 업데이트된) 데이터셋을 반환하고, 그걸 다시 klue_tc_train에 할당합니다.

klue_tc_train[0]
````

````python
# 학습/검증/테스트 데이터셋 분할
train_dataset   = klue_tc_train     .train_test_split(test_size =   10_000, shuffle = True, seed = 42)['test']
dataset         = klue_tc_eval      .train_test_split(test_size =    1_000, shuffle = True, seed = 42)
test_dataset    = dataset['test' ]
valid_dataset   = dataset['train']  .train_test_split(test_size =    1_000, shuffle = True, seed = 42)['test']
````

## 3.4.2 트레이너 API를 사용해 학습하기

````python
# Trainer를 사용한 학습 - (1) 준비
import torch
import numpy as np

from transformers import (
    Trainer,
    TrainingArguments,
    AutoModelForSequenceClassification,
    AutoTokenizer
)

def tokenize_function(examples):
    """
    데이터의 title 컬럼에 토큰화 수행
    ㅇ truncation=True
        - 입력이 max_length보다 길면 잘라냅니다(자르는 위치는 모델/토크나이저 규칙에 따름).
        - 반환값은 일반적으로 input_ids, attention_mask(및 토크나이저 설정에 따라 token_type_ids)를 포함한 dict입니다.
    """
    return tokenizer(examples['title'], padding = 'max_length', truncation = True)

model_id    = 'klue/roberta-base'
model       = (
    AutoModelForSequenceClassification
        .from_pretrained(
            model_id, # klue/roberta-base 바디
            num_labels = len(train_dataset.features['label'].names) # 데이터셋의 레이블 수 지정
        )
    )
tokenizer   = AutoTokenizer.from_pretrained(model_id)

train_dataset   = train_dataset .map(tokenize_function, batched = True)
valid_dataset   = valid_dataset .map(tokenize_function, batched = True)
test_dataset    = test_dataset  .map(tokenize_function, batched = True)
````

````python
# Trainer를 사용한 학습: (2) 학습 인자와 평가 함수 정의
training_args   = TrainingArguments(
    output_dir          = './results',
    num_train_epochs    = 1,
    per_device_train_batch_size = 8, #  배치 크기
    per_device_eval_batch_size  = 8,
    eval_strategy               = 'epoch', # --> 한 에포크 학습일 끝날 때마다 검증 데이터셋 평가 수행
    learning_rate               = 5e-5,
    push_to_hub                 = False,
    report_to = ["none"],   # W&B 비활성화 ??
)

def compute_metrics(eval_pred): # 학습 평가 지표 정의
    logits, labels  = eval_pred # 모델 예측 결과(eval_pred) 에서 logits 과 label unpack
    predictions     = np.argmax(logits, axis = -1) # logit이  가장 큰 값을 갖는 클래스를 뽑아서 predictions 에 저장
    return {'accuracy': (predictions == labels).mean()} # predictions와 labels 가 같은 값을 갖는 결과의 비율을 정확도로 저장
````

````python
# Trainer를 사용한 학습:(3) 학습 진행
trainer = Trainer(
    model   = model,
    args    = training_args,
    train_dataset   = train_dataset,
    eval_dataset    = valid_dataset,
    tokenizer       = tokenizer,
    compute_metrics = compute_metrics,
)

# trainer.train()

# trainer.evaluate(test_dataset)
````

### 데이터 준비
- make_dataloader 함수
    - 내부에서 tokenize_function 함수를 사용해 토큰화 수행
    - renae_column 메서드 : "label" --> "labels"로 변경
    - remove_columns : "title" 컬럼 제거

````python
def make_dataloader(dataset, batch_size, shuffle = True):
    dataset = dataset.map(tokenize_function, batched = True).with_format('torch')
    dataset = dataset.rename_column('label', 'labels')
    dataset = dataset.remove_columns(column_names = ['title'])
    return DataLoader(dataset, batch_size = batch_size, shuffle = shuffle)

# 데이터로더 만들기
train_dataloader    = make_dataloader(train_dataset,    batch_size = 8, shuffle = True  )
valid_dataloader    = make_dataloader(valid_dataset,    batch_size = 8, shuffle = False )
test_dataloader     = make_dataloader(test_dataset,     batch_size = 8, shuffle = False )
````

### 학습 함수 정의

````python
def train_epoch(model, data_loader, optimizer):
    model.train() # 모델을 학습 모드로 변환
    total_loss = 0

    for batch in tqdm(data_loader):
        optimizer.zero_grad()

        input_ids       = batch['input_ids'     ].to(device) # 모델에 입력할 토큰 아이디
        attention_mask  = batch['attention_mask'].to(device) # 모델에 입력할 어텐션 마스크
        labels          = batch['labels'        ].to(device) # 모델에 입력할 레이블

        # outputs         = model(input_ids, attention_mask, labels) # 모델 계산
        outputs         = model(input_ids = input_ids, attention_mask = attention_mask, labels = labels) # 모델 계산
        """
        model(input_ids, attention_mask, labels)처럼 위치(포지셔널) 인수로 labels를 전달한 것이 문제일 가능성이 큽니다.
        Hugging Face 모델의 호출 시 세 번째 위치 인수는 token_type_ids로 해석될 수 있어서(시퀀스 길이와 같은 shape를 기대) labels(예: 길이 8짜리 one-hot 또는 클래스 벡터)을 그 자리에 넣으면 seq_len(=512)과 맞지 않아 size 512 vs 8 오류가 발생합니다.
        """
        loss            = outputs.loss # 손실

        loss.backward() # 역전파
        optimizer.step() # 모델 업데이트
        total_loss += loss.item()

    avg_loss = total_loss / len(data_loader)
    return avg_loss
````

### 평가 함수 정의

````python
def evaluate(model, data_loader):
    model.eval() # 모델을 추론 모드로 변환
    total_loss  = 0
    predictions = []
    true_labels = []

    with torch.no_grad():
        for batch in tqdm(data_loader):
            optimizer.zero_grad()

            input_ids       = batch['input_ids'     ].to(device) # 모델에 입력할 토큰 아이디
            attention_mask  = batch['attention_mask'].to(device) # 모델에 입력할 어텐션 마스크
            labels          = batch['labels'        ].to(device) # 모델에 입력할 레이블

            # outputs         = model(input_ids, attention_mask, labels) # 모델 계산
            outputs         = model(input_ids = input_ids, attention_mask = attention_mask, labels = labels) # 모델 계산

            logits      = outputs.logits
            loss        = outputs.loss
            total_loss  += loss.item()

            preds       = torch.argmax(logits, dim = -1)
            predictions.extend(preds    .cpu().numpy())
            true_labels.extend(labels   .cpu().numpy())

    avg_loss    = total_loss / len(data_loader)
    accuracy    = np.mean(np.array(predictions) == np.array(true_labels))
    return avg_loss, accuracy
````

### 학습 수행

````python
num_epochs  = 1
optimizer   = AdamW(model.parameters(), lr = 5e-5)

# 학습 루프
for epoch in range(num_epochs):
    print(f'Epoch {epoch+1}/{num_epochs}')
    train_loss  = train_epoch(model, train_dataloader, optimizer)
    print(f'Training loss: {train_loss}')
    valid_loss, valid_accuracy = evaluate(model, valid_dataloader)
    print(f'Validation loss: {valid_loss}')
    print(f'Validation accurady: {valid_accuracy}')

# 테스트
_, test_accuracy = evaluate(model, test_dataloader)
print(f'Test accuracy: {test_accuracy}')
````
